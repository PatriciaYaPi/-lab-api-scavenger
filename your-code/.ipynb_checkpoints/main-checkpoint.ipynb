```python
data_pt_commits = requests.get('https://api.github.com/repos/ta-data-pt-rmt/lab-git/commits')
print("DATA RMTEU PART TIME:", data_pt_commits.status_code)
```

```python
data_pt_commits = pd.DataFrame.from_dict(data_pt_commits.json())
data_pt_commits.head()
```

```python
data_pt_commits['author'][0]
```

```python
commits_author_list = pd.json_normalize(data_pt_commits['author'])
commits_author_list.head()
```

```python
commits_list = pd.json_normalize(data_pt_commits['commit'])
commits_list.head()
```

```python
commits_list['author.date'] #.loc['2022-04-25':'2022-04-10']
```

```python
commits_list['author.date'] = pd.to_datetime(commits_list['author.date'], format='%Y-%m-%d')
#display(commits_list.dtypes)
commits_list.head()
```

```python
commits_list.sort_values(by='author.date', ascending=True)
```

```python
# No results within previous 15 days
commits_list[commits_list['author.date'] > '2022-06-23']
```
